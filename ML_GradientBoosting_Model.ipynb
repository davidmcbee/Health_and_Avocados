```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path
```

```python
cleaned_df = pd.read_csv(Path('../resources/combined_cleaned.csv'))
cleaned_df.head()
```

```python
cleaned_df = cleaned_df.drop(columns=['Unnamed: 0', 'index'], axis=1)
cleaned_df.head()
```

```python
price_bins = [0.44, 1.14, 1.37, 1.63, 3.17]
group_names = ["0", "1", "2", "3" ]

cleaned_cat_df = cleaned_df.copy()

# Categorize prices for NB.
cleaned_cat_df["price_cat"] = pd.cut(cleaned_cat_df['avg_price'], price_bins, labels=group_names)

cleaned_cat_df.head()
```

```python
def GBSSModel4(X, y):
    # Imports
    from sklearn.model_selection import train_test_split
    from sklearn.preprocessing import StandardScaler
    # Gradient Booster modifier
    from sklearn.ensemble import GradientBoostingClassifier
    # additional imports
    from sklearn.metrics import confusion_matrix
    from sklearn.metrics import accuracy_score
    from sklearn.metrics import classification_report
    
    # Splitting model
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=0.70)
    
    # Standard Scaler
    scaler = StandardScaler()
    X_scaler = scaler.fit(X_train)
    X_train_scaled = X_scaler.transform(X_train)
    X_test_scaled = X_scaler.transform(X_test)
    
    learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
    
    # learning rates and accuracy scores
    #for learning_rate in learning_rates:
    #    classifier = GradientBoostingClassifier(n_estimators=1024,
    #                                            learning_rate=learning_rate,
    #                                            max_features=264,
    #                                            max_depth=5,
    #                                            random_state=0)
    #    
    #    classifier.fit(X_train_scaled, y_train)
    #    
    #    print("Learning rate: ", learning_rate)
    #    print("Accuracy score (training): {0:.3f}".format(classifier.score(X_train_scaled, y_train)))
    #    print("Accuracy score (validation): {0:.3f}".format(classifier.score(X_test_scaled, y_test)))
    #    
    # Classifier for gradient boosting
    classifier = GradientBoostingClassifier(n_estimators=1024,
                                            learning_rate=0.5, 
                                            max_features=264, 
                                            max_depth=5, 
                                            random_state=0)
    
    # Fitting classifier and predictions
    classifier.fit(X_train_scaled, y_train)
    predictions = classifier.predict(X_test_scaled)
    
    acc_score = accuracy_score(y_test, predictions)
    print(f"Accuracy Score : {acc_score}")
    
    # Confusion matrix
    cm = confusion_matrix(y_test, predictions)
    cm_df = pd.DataFrame(cm, index=["Actual 0", "Actual 1", "Actual 2", "Actual 3"], 
                         columns=["Predicted 0", "Predicted 1", "Predicted 2", "Predicted 3"])
    display(cm_df)
    
    # Classification report
    print("Classification Report")
    print(classification_report(y_test, predictions))
    
```
