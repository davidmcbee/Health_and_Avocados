```python
# import dependencies
import pandas as pd
import datetime
```

# AvoProduction_2018

```python
# read in 2018 avo prod
prod_2018 = pd.read_csv("raw/AvoProduction_2018.csv")
prod_2018.head()
```

```python
# rename columns
prod_2018.columns = ["date", "status", "total_volume", "california", "chile", "mexico", "peru", "columbia"]

# change status to lowercase
prod_2018.loc[:,'status'] = prod_2018.loc[:, 'status'].str.lower()

cols = list(prod_2018.columns.values)
cols
```

```python
# add year_month column and move to first column
prod_2018['year_month'] = prod_2018['date']
prod_2018 = prod_2018[["year_month",
                         'date',
                         'status',
                         'total_volume',
                         'california',
                         'chile',
                         'mexico',
                         'peru',
                         'columbia']]
prod_2018.head()
```

```python
# check column dtypes
prod_2018.dtypes
```

```python
# change date to datetime format
prod_2018[['year_month', "date"]] = prod_2018[['year_month', "date"]].apply(pd.to_datetime)
prod_2018.dtypes
```

```python
# change year_month format
# prices_2018["year_month"] = prices_2018.date.dt.strftime('%Y%m').astype(object)
prod_2018["year_month"] = pd.to_datetime(prod_2018["year_month"]).dt.to_period("m")
prod_2018
```

```python
# check column dtypes
prod_2018.dtypes
```

```python
# export prod_2018
prod_2018.to_csv('resources/production_2018.csv', index=False)
```
