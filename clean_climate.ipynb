```python
# import dependencies
import pandas as pd
import datetime
```

```python
# read in climate data
climate = pd.read_csv("raw/CA_State_Climate.csv")
climate.head()
```

```python
# drop unnecessary columns
climate = climate.drop(columns=["StateCode", "Division"])
climate.head()
climate.dtypes
```

```python
# change YearMonth format
climate["YearMonth"] = pd.to_datetime(climate["YearMonth"]).dt.to_period("m")
climate
```

```python
# check column dtypes
climate.dtypes
```

```python
# export climate
climate.to_csv('resources/climate.csv', index=False)
```
