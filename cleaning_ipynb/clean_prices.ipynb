```python
# import dependencies
import pandas as pd
import datetime
```

# AvocadoBoard_2017

```python
# read in 2017 avo prices
prices_2017 = pd.read_csv("raw/AvocadoBoard_2017.csv")
prices_2017.head()
```

```python
# rename columns
prices_2017.columns = ["geography", "timeframe", "date", "type", "avg_price", "prices_total_volume", "units_4046",
                      "units_4225", "units_4770", "total_bags", "s_bags", "l_bags", "xl_bags"]

# change type to lowercase
#prices_2017["type"] = prices_2017["type"].str.lower()
prices_2017.loc[:,'type'] = prices_2017.loc[:, 'type'].str.lower()
prices_2017.loc[:,'timeframe'] = prices_2017.loc[:, 'timeframe'].str.lower()

cols = list(prices_2017.columns.values)
cols
```

```python
# add year_month column and move to first column
prices_2017['year_month'] = prices_2017['date']
prices_2017 = prices_2017[['year_month',
                             'geography',
                             'timeframe',
                             'date',
                             'type',
                             'avg_price',
                             'prices_total_volume',
                             'units_4046',
                             'units_4225',
                             'units_4770',
                             'total_bags',
                             's_bags',
                             'l_bags',
                             'xl_bags']]
prices_2017.head()
```

```python
# remove spaces and dots
prices_2017.geography = prices_2017.geography.str.replace(' ', '')
prices_2017.geography = prices_2017.geography.str.replace('.', '')
#prices_2017.geography = prices_2017.geography.str.replace('/', '_')
prices_2017
```

```python
prices_2017.geography.unique()
```

```python
# check column dtypes
prices_2017.dtypes
```

```python
# change date to datetime format
prices_2017[['year_month', "date"]] = prices_2017[['year_month', "date"]].apply(pd.to_datetime)
prices_2017.dtypes
```

```python
# change year_month format
#prices_2017["year_month"] = prices_2017.date.dt.strftime('%Y%m').astype(object)
#prices_2017["year_month"] = pd.to_datetime(prices_2017["year_month"]).dt.to_period("m")
#prices_2017
```

```python
# year_month to include -01
prices_2017.year_month = prices_2017.year_month.apply(lambda dt: dt.replace(day=1))
prices_2017.head()
```

```python
# check column dtypes
prices_2017.dtypes
```

```python
# export prices_2018
prices_2017.to_csv('resources/prices_2017.csv', index=False)
```
