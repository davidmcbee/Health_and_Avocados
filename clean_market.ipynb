```python
# import dependencies
import pandas as pd
import datetime
```

# 2017-market-data

```python
# read in 2017 market prices
market_2017 = pd.read_csv("raw/2017-market-region-hab-data.csv")
market_2017.head()
```

```python
# rename columns
market_2017.columns = ["geography", "segment", "variety", "timeframe", "period", "date", "units_prior_year",
                       "units_current_year", "unit_variance", "dollars_prior_year", "dollars_current_year",
                       "dollar_variance", "avg_price_prior_year", "avg_price_current_year", "avg_price_variance"]

# change type to lowercase
market_2017.loc[:,"segment"] = market_2017.loc[:,"segment"].str.lower()
market_2017.loc[:,"variety"] = market_2017.loc[:,"variety"].str.lower()
market_2017.loc[:,"timeframe"] = market_2017.loc[:,"timeframe"].str.lower()

cols = list(market_2017.columns.values)
cols
```

```python
# add year_month column and move to first column
market_2017['year_month'] = market_2017['date']
market_2017 = market_2017[['year_month',
                             'geography',
                             'segment',
                             'variety',
                             'timeframe',
                             'period',
                             'date',
                             'units_prior_year',
                             'units_current_year',
                             'unit_variance',
                             'dollars_prior_year',
                             'dollars_current_year',
                             'dollar_variance',
                             'avg_price_prior_year',
                             'avg_price_current_year',
                             'avg_price_variance']]
market_2017.head()
```

```python
# check column dtypes
market_2017.dtypes
```

```python
# change date to datetime format
market_2017[['year_month', "date"]] = market_2017[['year_month', "date"]].apply(pd.to_datetime)
market_2017.dtypes
```

```python
# change year_month format
market_2017["year_month"] = pd.to_datetime(market_2017["year_month"]).dt.to_period("m")
market_2017
```

```python
# check column dtypes
market_2017.dtypes
```

```python
# export market_2017
market_2017.to_csv('resources/market_2017.csv', index=False)
```
